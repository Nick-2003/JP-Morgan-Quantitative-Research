## Final Function
- Data after this section was for testing.

```python
def ficoratemap(ficodata, k): 
    """
    Use ficodata to create k ratings in terms of FICO scores, with descending order starting from 0. 
    
    Parameters:
    -----------
        
    ficodata: dataframe
        Dataframe of FICO data. 

    k: int
        Number of buckets. 
        
    Returns:
    --------
    New dataframe with corresponding ratings 
    """ 
    kmeans = KMeans(n_clusters = 10, n_init='auto', random_state = 44)
    kmeans.fit(ficodata["fico_score"].values.reshape(-1, 1))

    cluster_labels = kmeans.labels_
    data_points = kmeans.cluster_centers_
    ficodata["labels"] = cluster_labels
    
    # Initialize dictionaries to store maximum and minimum values for each label
    intervals = []

    # Iterate over data points and cluster labels to find maximum and minimum values for each label
    for label in np.unique(cluster_labels):
        interest = ficodata.loc[ficodata["labels"] == label]
        max_values = np.amax(interest["fico_score"])
        min_values = np.amin(interest["fico_score"])
        interval = pd.Interval(left = min_values, right = max_values, closed = "both")
        intervals.append(interval)

    def convertlist(intervallist):
        finallist = sorted(intervals, reverse = True)
        bigmin = finallist[0].left
        smallmax = finallist[-1].right
        
        newend = pd.Interval(left = bigmin, right = 850, closed = "both")
        newstart = pd.Interval(left = 300, right = smallmax, closed = "both")

        finallist.pop(0)
        finallist.pop(-1)
        finallist.insert(0, newend)
        finallist.insert(-1, newstart)
        return sorted(finallist, reverse = True)

    sorted_intervals = convertlist(intervals)

    def scoretolabel(x): 
        for interval in sorted_intervals: 
            if x in interval: 
                return sorted_intervals.index(interval)

    ficodata["sort_label"] = ficodata["fico_score"].apply(lambda x: scoretolabel(x))
    print(sorted_intervals)
    return ficodata
```

```python
newdata = ficoratemap(train_df, 10)
display(newdata.head())
```

## Final Function (After the above testing)

```python
def ficoratemap(ficodata, k): 
    """
    Use ficodata to create k ratings in terms of FICO scores, with descending order starting from 0. 
    
    Parameters:
    -----------
        
    ficodata: dataframe
        Dataframe of FICO data. 

    k: int
        Number of buckets. 
        
    Returns:
    --------
    New dataframe with corresponding ratings 
    """ 
    kmeans = KMeans(n_clusters = 10, n_init='auto', random_state = 44)
    kmeans.fit(ficodata["fico_score"].values.reshape(-1, 1))

    cluster_labels = kmeans.labels_
    data_points = kmeans.cluster_centers_
    ficodata["labels"] = cluster_labels
    
    # Initialize dictionaries to store maximum and minimum values for each label
    intervals = []

    # Iterate over data points and cluster labels to find maximum and minimum values for each label
    for label in np.unique(cluster_labels):
        interest = ficodata.loc[ficodata["labels"] == label]
        max_values = np.amax(interest["fico_score"])
        min_values = np.amin(interest["fico_score"])
        interval = pd.Interval(left = min_values, right = max_values, closed = "both")
        intervals.append(interval)

    def convertlist(intervallist):
        finallist = sorted(intervals, reverse = True)
        bigmin = finallist[0].left
        smallmax = finallist[-1].right
        
        newend = pd.Interval(left = bigmin, right = 850, closed = "both")
        newstart = pd.Interval(left = 300, right = smallmax, closed = "both")

        finallist.pop(0)
        finallist.pop(-1)
        finallist.insert(0, newend)
        finallist.insert(-1, newstart)
        return sorted(finallist, reverse = True)

    sorted_intervals = convertlist(intervals)

    def scoretolabel(x): 
        for interval in sorted_intervals: 
            if x in interval: 
                return sorted_intervals.index(interval)

    ficodata["sort_label"] = ficodata["fico_score"].apply(lambda x: scoretolabel(x))
    print(sorted_intervals)
    return ficodata
```
